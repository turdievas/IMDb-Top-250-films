```python
import schedule
import time
import pandas as pd

# Load data
def load_data():
    
    df = pd.read_excel("movies.xlsx")
    return df

# Compute KPIs
def compute_kpis():
    df = load_data()
    print(f"Weekly KPIs:")

    #Number of movies
    #1. 
    num_rows = df.shape[0]
    num_rows
    print(f"The DataFrame has {num_rows} rows.")
    summary_df = pd.DataFrame({
        "Metric": ["Total Rows"],
        "Value": [num_rows]
    })

    summary_df.style.format({'Value': '{:,}'})

    # Average rating
    #2.
    average_rating = df['Rating'].mean()
    average_rating
    print(f"The average rate is {average_rating}.")
    avg_rate = pd.DataFrame({
        "Metric": ["Average Rating"],
        "Value": [average_rating]
    })

    avg_rate.style.format({'Value': '{:,}'}) 

    #Total viewers 
    #3.
    total_viewers = df['Viewers_num'].sum()
    total_viewers
    print(f"The total viewers: {total_viewers}.")
    tt_viewers = pd.DataFrame({
        "Metric": ["Total Viewers"],
        "Value": [total_viewers]
    })

    tt_viewers.style.format({'Value': '{:,}'})

    #Average duration
    #4.
    avg_duration  = df['DurationMinutes'].mean()
    avg_duration
    print(f"The average duration in minutes: {avg_duration}.")
    avg_dur = pd.DataFrame({
        "Metric": ["Average Duration"],
        "Value": [avg_duration]
    })

    avg_dur.style.format({'Value': '{:,}'})

    #Which year brought the most views?
    #5. 
    #Movie KPIs
    viewers_by_year = df.groupby('Year')['Viewers_num'].sum().sort_values(ascending=False)
    print("Total viewers by year (sorted):\n", viewers_by_year, "\n")
    print(f"The year with the most viewers is {viewers_by_year.index[0]} with {viewers_by_year.iloc[0]} viewers.\n")

    

    
    # FIle saving
    with open("kpis.txt", "a") as f:
        f.write(f"""\nNumber of movies: {summary_df} \nAverage Rating: {avg_rate} 
                    \nTotal viewers: {tt_viewers.style.format({'Value': '{:,}'})}
                    \nAverage duration: {avg_dur.style.format({'Value': '{:,}'})}
                    \nTotal viewers by year (sorted):\n", {viewers_by_year}
                    \n"The year with the most viewers is {viewers_by_year.index[0]} with {viewers_by_year.iloc[0]} viewers """)

# Scheduling
schedule.every().monday.at("09:00").do(compute_kpis)  # runs every Monday 9 AM

# Running every Monday
while True:
    schedule.run_pending()
    time.sleep(60)
```
