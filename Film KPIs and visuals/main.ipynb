```python
#Movie KPIs
import pandas as pd
import openpyxl

data = pd.read_excel("Book1.xlsx")
df = pd.DataFrame(data)
df
styled = (df.style
            .background_gradient(subset=['Rating'], cmap='YlOrBr')
            .format({'Viewers': '{:,}'}))
styled
```

```python
#Number of movies
#1. 

num_rows = df.shape[0]
num_rows
print(f"The DataFrame has {num_rows} rows.")
summary_df = pd.DataFrame({
    "Metric": ["Total Rows"],
    "Value": [num_rows]
})

summary_df.style.format({'Value': '{:,}'})
```

```python
# Average rating
#2.
average_rating = df['Rating'].mean()
average_rating
print(f"The average rate is {average_rating}.")
avg_rate = pd.DataFrame({
    "Metric": ["Average Rating"],
    "Value": [average_rating]
})

avg_rate.style.format({'Value': '{:,}'})
```

```python
#Total viewers 
#3.
total_viewers = df['ViewersCount'].sum()
total_viewers
print(f"The total viewers: {total_viewers}.")
tt_viewers = pd.DataFrame({
    "Metric": ["Total Viewers"],
    "Value": [total_viewers]
})

tt_viewers.style.format({'Value': '{:,}'})
```

```python
#Average duration
#4.
avg_duration  = df['DurationHours'].mean()
avg_duration
print(f"The average duration: {avg_duration}.")
avg_dur = pd.DataFrame({
    "Metric": ["Average Duration"],
    "Value": [avg_duration]
})

avg_dur.style.format({'Value': '{:,}'})
```

```python
#Which year brought the most views?
#5. 
#Movie KPIs
import pandas as pd
import openpyxl

data = pd.read_excel("Book1.xlsx")
df = pd.DataFrame(data)
df
styled = (df.style
            .background_gradient(subset=['Rating'], cmap='YlOrBr')
            .format({'Viewers': '{:,}'}))
styled
viewers_by_year = df.groupby('Year')['ViewersCount'].sum().sort_values(ascending=False)
print("Total viewers by year (sorted):\n", viewers_by_year, "\n")
print(f"The year with the most viewers is {viewers_by_year.index[0]} with {viewers_by_year.iloc[0]} viewers.\n")
```

```python
# Did Higher-Rated Films Get More Viewers?
#6.
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Rating'], df['ViewersCount'], alpha=0.7)
plt.title('Relationship Between Rating and Viewer Count')
plt.xlabel('Rating')
plt.ylabel('Viewer Count')
plt.grid(True)
plt.show()

print("The scatter plot shows the relationship between ratings and viewer counts.")
```

```python
#The most popular film
#7.
most_popular_film = df.loc[df['ViewersCount'].idxmax()]
print(f"The most popular film : {most_popular_film}")
```

```python
#Do Longer Films Tend to Have Higher Ratings? 
#8
plt.figure(figsize=(10, 6))
plt.scatter(df['DurationMinutes'], df['Rating'], alpha=0.7)
plt.title('Relationship Between Duration and Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
print(f"The graph shows the relationship between duration and rating.")
```
